# Import Library

```python
#Import Library
import pandas as pd
import numpy as np
import joblib
import json
```

# Data Loading

```python
# Data Loading
inference = pd.read_csv('inference.csv')
inference.info()
```

# Model Inference

```python
# Model Inference
model = joblib.load('model.pkl')
with open('feature.json','r') as file:
    feature = json.load(file)
```

```python
# Model Inference
inference.insert(0,'predict',model.predict(inference[feature]))
inference.head()
```
